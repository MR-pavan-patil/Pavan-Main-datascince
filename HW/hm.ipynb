```python
def grade(x):
    if pd.isna(x):
        return "Not Available"
    elif x >= 9.0:
        return "Excellent"
    elif x >= 8.0:
        return "Very Good"
    elif x >= 7.0:
        return "Good"
    elif x >= 6.0:
        return "Average"
    elif x < 6.0:
        return "Below Average"


    
```

```python
df["Grade"]=df["cgpa"].apply(grade)
```
