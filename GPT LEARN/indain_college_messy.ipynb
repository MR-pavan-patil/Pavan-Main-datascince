## cleaning the hostel column

```python
df["hostel"].unique()
```

```python
df["hostel"]=df["hostel"].apply(lambda x: x.upper())
```

```python
df["hostel"].unique()
```

```python
def hostel_clean(x):
    if x == "N":
        x=x.replace("N","NO")
        return x
    else:
        x= "Yes"
        return x
```

```python
df["hostel"]=df["hostel"].apply(hostel_clean)
```

```python
df["hostel"].unique()
```

#### cleaning state data

```python
df["state"] = df["state"].apply(lambda x: x.title().strip())
```

```python
df["state"].unique()
```

```python
def state_1(x):
    if x in ["Tn"]:
       return "Tamil Nadu"
    elif x in ["Rj"]:
        return "Rajasthan"
    elif x in ["Dl"]:
        return "Delhi"
    elif x in "Ts":
        return "Telangana"
    elif x in "Wb":
        return "West Bengal"
    elif x in "Mh":
        return "Maharashtra"
    elif x in "Gj":
        return  "Gujarat"
    elif x in "Up":
        return "Uttar Pradesh"
    elif x in "Ka":
        return "Karnataka"
    elif x in "Kl":
        return "Kerala"
    else:
        return x



```

```python
df["state"]=df["state"].apply(state_1)
```

#### cleaning Gender colunm

```python
df["gender"].unique()
```

```python
def gender_clean(y):
    x= str(y).upper()
    if x in ["M","MALE","MAN"]:
        return "Male"
    elif x in ["F", "WOMAN"]:
        return "Female"
    else:
        return "OTHER"
```

```python
df["gender"]=df["gender"].apply(gender_clean)
```
