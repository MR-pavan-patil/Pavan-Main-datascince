```python
import pandas as pd

data = {
    'Department': ['IT','HR','IT','HR','Finance','IT','Finance'],
    'Employee':  ['Pavan','Ravi','Anu','Sneha','Arjun','Riya','Divya'],
    'Salary':    [50000,45000,52000,48000,55000,53000,60000],
    'Experience':[2,3,4,5,3,2,6]
}

data = pd.DataFrame(data)
```

```python
data.head(10)
```

```python
data["Department"].unique()
```

```python
data.groupby("Department")["Salary"].mean()
```

```python
data.groupby("Department")["Experience"].max()
```

```python
data.groupby("Department")["Experience"].min()
```

```python
data.groupby("Department")[["Salary","Experience"]].mean().max()
```

```python
data.groupby("Department")[["Salary","Experience"]].agg(['mean','max','min','std'])
```

```python
data.groupby("Department").agg(
    av_salary = ('Salary','mean')
)
```
