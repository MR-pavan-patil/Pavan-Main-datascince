```python
def biggestnum():
    a=int(input("enter a number"))
    b=int(input("enter a number"))
    c=int(input("enter a number"))
    print(f"numbers:- {a} {b} {c}")
    if a > b and a > c:
        print(f"{a} is biggest number")
    elif b >= a and  b >= c:
        print(f"{b} is biggest number ")
    else:
        print(f"{c} is biggest number")
```

```python
biggestnum()
```
